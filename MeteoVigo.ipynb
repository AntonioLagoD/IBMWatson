```python
from statistics import mean
```

```python
with open('temperaturas2019.txt',"r") as file1:
    with open('medias2019.txt','w') as file2:
        for line in file1:
          mes = line.split(',')
          temps = list(map(float, mes[1:]))
          mediaMes=mean(temps)
          linea = mes[0] + ',' +str(mediaMes)+'\n'
          file2.write(linea)
          
```

```python
import numpy as np
import matplotlib.pyplot as plt
archivo = "medias2019.txt"
data = np.genfromtxt(archivo, delimiter=',', dtype=None, skip_header=0, names=('meses', 'temperatura'),encoding=None)
print(data)
plt.figure(figsize=(14,8))
plt.title("Temperatura media por mes en Vigo - 2019")
plt.xlabel("Meses")
plt.ylabel("ºC")
plt.plot(data['meses'], data['temperatura'], color='orange')
plt.legend(['T medias 2019'], loc='upper left')
plt.show()
```
